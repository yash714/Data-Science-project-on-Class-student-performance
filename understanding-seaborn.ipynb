```python
df= pd.read_csv('StudentsPerformance.csv')
```

# import seaborn 

```python
import seaborn as sns
```

```python
# display the first 4 rows
df.head()
```

# Visualization

### gender ratio

```python
sns.countplot(x='gender', data=df)
```

## parent's education

```python
plt.figure(figsize=(12, 7))
sns.countplot(x='parental level of education', data=df, palette='flare')
```

## Ethinicity

```python
plt.figure(figsize=(12, 7))
sns.countplot(x='race/ethnicity', data=df, palette='flare')
```

### based on lunch segmentation

```python
sns.countplot(x='lunch', data=df)
```

# Test preparation course

```python
sns.countplot(x='test preparation course', data=df)
```

# Categorical Plots - Distribution within Categories

## Math score


```python
sns.barplot(x='gender', y='math score', data=df)
```

### understand the score based on parent's education

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='race/ethnicity', y='math score', data=df, hue='parental level of education', palette='pastel');

plt.legend(loc=(1.05, 0.5))
```

## visualize math score data with test preparation course.

```python
plt.figure(figsize=(25, 6), dpi=200)
sns.countplot(x='math score', data=df, hue='test preparation course')
```

## boxplot() distribution

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='math score', data=df)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='math score', data=df, hue='gender')
```

# Comparison Plots with pairplot() and jointplot()

### Relationship between different subject scores with gender segmentation

```python
sns.pairplot(df, hue='gender', palette='viridis');
```

```python
sns.jointplot(x='math score', y='reading score', data=df, alpha=0.5, kind='hex')
```
