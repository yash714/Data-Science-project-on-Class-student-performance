```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("StudentsPerformance.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.corr(), annot = True, fmt='.2g', cmap='YlGnBu', linewidths=1, linecolor='black',square=True)
```

```python
matrix = np.triu(data.corr())
sns.heatmap(data.corr(), annot=True, linewidths=1, mask=matrix)
```

```python
data_1 = data.copy()
```

```python
def rename(data):
  for label,content in data.items():
    if not pd.api.types.is_numeric_dtype(content):
      data[label] = pd.Categorical(content).codes+1
  return data
```

```python
transformed_data = rename(data_1)
transformed_data.head()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(transformed_data.corr(), annot = True, fmt='.2g',  linewidths=1, linecolor='black',square=True, cmap='coolwarm')
```

```python
plt.figure(figsize=(10, 8))
matrix = np.triu(transformed_data.corr())
sns.heatmap(transformed_data.corr(), annot=True, linewidths=1, mask=matrix, cmap='coolwarm')
```
